```python
import matplotlib.pyplot as plt
import numpy as np
import tensor


%matplotlib inline
```

```python
def sigmoid(z):
    
    return 1/(1 + np.exp(-z))
```

```python
sample_z = np.linspace(-10,10,100)
sample_a = sigmoid(sample_z)
```

```python
plt.plot(sample_z, sample_a)
```

```python
g = tensor.Graph()
g.set_as_default()
x = tensor.Placeholder()
w = tensor.Variable([1,1])
b = tensor.Variable(-5)
z = tensor.add(tensor.matmul(w,x),b)
```

```python
a = tensor.Sigmoid(z)
```

```python
sess = tensor.Session()
sess.run(operation=a, feed_dict={x:[2,-10]})
```
